# Add Two Ints

```python
def handle_add_two_ints(req):
    print(f"Returning [{req.a} + {req.b} = {req.a + req.b}]")
    return AddTwoIntsResponse(req.a + req.b)
```

```python
%%thread_cell

srv = rospy.Service('add_two_ints', AddTwoInts, handle_add_two_ints)
```

```python
jupyros.service_client('add_two_ints', AddTwoInts)
```
